```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# Generating some random points
points = np.array([
    [np.random.uniform(), np.random.uniform()]
    for i in range(10)
])

# Compute Voronoi
vor = Voronoi(points)

# Plot
voronoi_plot_2d(vor)

# Mark the input points
plt.plot(points[:, 0], points[:, 1], 'ro')

plt.xlim([0, 1])
plt.ylim([0, 1])
plt.show()
```
